```python
avocado_df = avocado_df.sort_values('Date')
avocado_df
```

```python
plt.figure(figsize = (10, 10))
plt.plot(avocado_df['Date'], avocado_df['AveragePrice'])
plt.show()
```

```python
plt.figure(figsize = [25, 12])
palette = sns.color_palette("husl", len(avocado_df['region'].unique()))
sns.countplot(x = 'region', data = avocado_df, palette = palette)
plt.xticks(rotation = 80)
plt.title("Avocado Sales by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```

```python
palette = sns.color_palette("magma", len(avocado_df['year'].unique()))
sns.countplot(x = 'year', data = avocado_df, palette = palette)
plt.xticks(rotation = 80)
plt.title("Avocado Sales by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
avocado_prophet_df = avocado_df[['Date', 'AveragePrice']]
avocado_prophet_df
```

```python
avocado_prophet_df = avocado_prophet_df.rename(columns={'Date':'ds', 'AveragePrice':'y'})
```

```python
avocado_prophet_df
```

```python
m = Prophet()
```

```python
m.fit(avocado_prophet_df)
```

```python
#forecast the future

future = m.make_future_dataframe(periods=365)
forecast = m.predict (future)
```

```python
forecast
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure = m.plot_components(forecast)
```
