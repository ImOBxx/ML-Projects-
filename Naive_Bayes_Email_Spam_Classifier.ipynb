```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
spamham_countvectorizer = vectorizer.fit_transform(spam_df['text'])
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(spamham_countvectorizer.toarray())
```

```python
spamham_countvectorizer.shape
```

```python
label = spam_df['spam'].values

from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()

```

```python
label
```

```python
NB_classifier.fit(spamham_countvectorizer, label)
```

```python
test_sample = ['Free Money!!!', 'Hi Kim, let me know if need any further information.']
test_sample_countvectorizer = vectorizer.transform(test_sample)
```

```python
test_predict = NB_classifier.predict(test_sample_countvectorizer)
test_predict
```

```python
test_sample = ['Hello, I am Ryan, I would like to book a hotel in Bali by January 24th', 'money!!!!viagra']
test_sample_countvectorizer = vectorizer.transform(test_sample)
test_predict = NB_classifier.predict(test_sample_countvectorizer)
test_predict
```

```python
X = spamham_countvectorizer
y = label
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = NB_classifier.predict(X_train)
```

```python
y_predict_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot = True)
```
